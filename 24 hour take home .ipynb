```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('customers_data.csv')
df.head()
```

```python
df.info()
```

```python
df['gender'].replace(['male','female'],[1,0], inplace=True)
df['card_on_file'].replace(['no','yes'],[0,1], inplace=True)
df['loyalty'].replace([False,True],[0,1], inplace=True)
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
corr = df.corr()
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, annot = True, annot_kws={'size':12})
heat_map=plt.gcf()
heat_map.set_size_inches(20,15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
y = df['loyalty']
x = df.drop('loyalty', axis=1)
x
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.25,random_state=99)
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(x_train,y_train)
nb_y_pred = bnb.predict(x_test)

```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, nb_y_pred))
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.25,random_state=99)
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
```

```python
logisticRegr = LogisticRegression(penalty='l1')
logisticRegr.fit(X=x_train, y=y_train)
test_y_pred = logisticRegr.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,test_y_pred))
```

```python
print( 'Score: ' , logisticRegr.score(x_test, y_test))
```

```python
from sklearn.svm import SVC
svm = SVC(kernel = 'linear')
svm.fit(x_train,y_train)
svm_predict = svm.predict(x_test)
```

```python
metrics.accuracy_score(y_test, svm_predict)
```

```python
print( 'Score: ' , svm.score(x_test, y_test))
```

```python
from sklearn import tree


decision_tree = tree.DecisionTreeClassifier(
    criterion='entropy',
    max_features=1,
    max_depth=4,
    random_state = 1337
)
decision_tree.fit(x_train, y_train)
```

```python
dt_predicted = decision_tree.predict(x_test)
```

```python
decision_tree.score(x_test, y_test)
```
